```python
# Average order value for each customer

order_summary = df_copy.groupby('Order Number').agg({
    'Sales Amount': 'sum',
    'Sales Quantity': 'sum'
}).reset_index()

# Merge the order summary back to the original DataFrame
df_copy = pd.merge(df_copy, order_summary, on='Order Number', suffixes=('', '_total'))

# Calculate the average order value for each customer
df_copy['Average Order Value'] = df_copy['Sales Amount'] / df_copy['Sales Quantity']
# Display the results (optional)
df_copy[['CustKey', 'Order Number', 'Sales Amount', 'Sales Quantity', 'Average Order Value']].head(10)
```

```python


# Create a DataFrame of unique customers and their first purchase date
first_purchase_dates = df.groupby('CustKey')['Invoice Date'].min().reset_index()
first_purchase_dates.columns = ['CustKey', 'FirstPurchaseDate']

# Merge the first purchase dates back into the original DataFrame
df = pd.merge(df, first_purchase_dates, on='CustKey')

# Calculate the time since the first purchase for each transaction
df['DaysSinceFirstPurchase'] = (df['Invoice Date'] - df['FirstPurchaseDate']).dt.days

# Define time periods for analysis (e.g., monthly)
time_periods = [30, 60, 90, 180, 365]

# Calculate customer retention rates for each time period
retention_rates = []
for period in time_periods:
    retained_customers = df[df['DaysSinceFirstPurchase'] <= period]['CustKey'].nunique()
    total_customers = df[df['DaysSinceFirstPurchase'] == 0]['CustKey'].nunique()
    retention_rate = (retained_customers / total_customers) * 100
    retention_rates.append(retention_rate)

# Create a bar plot to visualize retention rates over time
plt.figure(figsize=(10, 6))
sns.barplot(x=time_periods, y=retention_rates, palette='Blues')
plt.title('Customer Retention Rates Over Time')
plt.xlabel('Time Period (Days)')
plt.ylabel('Retention Rate (%)')
plt.show()
```
